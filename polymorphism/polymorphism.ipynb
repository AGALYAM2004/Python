```python
class employee:
    def work(self):
        return ("working")
class developer(employee):
    def work(self):
        return ("Developing")
class HR(employee):
    def work(self):
        return ("recruiting")
class Data_Analyst(employee):
    def work(self):
        return ("Analyse")
    
employee=[developer(),HR(),Data_Analyst()]
for emp in employee:
    print(emp.work())
```
