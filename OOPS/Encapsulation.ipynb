```python
class employee:
    def __init__(self):
        self.__salary=50000
    def update_salary(self,salary):
     if salary>0:
        self.__salary=salary
        return ("Salary updated")
     else:
        return("Salary Invalid")
    def get_salary(self):
        return(self.__salary)
    
employee1=employee()
print(employee1.get_salary())
print(employee1.update_salary(10))
print(employee1.get_salary())
   
```
