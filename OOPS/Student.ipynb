```python
class student:
    def __init__(self,name,rollno):
        self.name=name
        self.rollno=rollno
        self.branch="ECE"
        

    def tell_details(self):
        return(f"My name is {self.name},My rollno is {self.rollno} and My branch is {self.branch}")

student1=student("Ram","12")
print(student1.branch)
print(student1.tell_details())
```

```python
class student:
    def __init__(self,age):
    
        self.age=age+10
    def tell_age(self):
        return(f"My age is {self.age}")
student1=student(20)
print(student1.age)
print(student1.tell_age())
```
